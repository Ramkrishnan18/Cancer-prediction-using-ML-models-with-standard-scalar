```python
import pandas as pd
df=pd.read_csv("/content/Cancer_Data.csv")
df
```

```python
df.isnull().sum()
```

```python
df.drop(["Unnamed: 32"],axis=1)
```

```python
x=df.iloc[:,2:32].values
x
```

```python
y=df["diagnosis"].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(x_train)
x_train=sc.transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
```

```python
y_predict=knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_predict)#knn
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_predictions(y_test,y_predict)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_predict))
```

```python
from sklearn.svm import SVC
svm=SVC(kernel="linear")
svm.fit(x_train,y_train)
```

```python
y_predict=svm.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_predict)#svm
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_predictions(y_test,y_predict)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_predict))
```

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(x_train,y_train)
```

```python
y_predict=nb.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_predict)#GaussianNB
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_predictions(y_test,y_predict)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_predict))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
y_predict=dt.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_predict)#DecisionTreeClassifier
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_predictions(y_test,y_predict)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_predict))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
y_predict=rf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_predict)#RandomForestClassifier
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_predictions(y_test,y_predict)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_predict))
```
